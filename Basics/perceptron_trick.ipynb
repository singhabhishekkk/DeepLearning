```python
from sklearn.datasets import make_classification
import numpy as np

# generating sample classification data
X, y = make_classification(n_samples=100, n_features=2,
                           n_informative=1, n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, 
                           random_state=41, hypercube=False, class_sep=10)
print("X is",X)
print("y is",y)
```

```python
np.insert(X,0,1,axis = 1)
```

```python
def perceptron(x,y):
    X = np.insert(x,0,1,axis = 1) # adding 1 in x (perceptron trick - bias theory)
    weights = np.ones(X.shape[1]) # initialising weights as 1
    lr = 0.1 # learning rate
    
    for i in range (100):
        j = np.random.randint(0,100) # since shape of x = (100,2) => selecting a random value between 0 to 100
        y_hat = step(np.dot(X[j],weights))  # predicting using initialised weights
        weights = weights + lr*(y[j]-y_hat)*X[j] # applying perceptron trick here, updating weights ( the line )
    return weights[0],weights[1:] # returning updated weights
    
    
    
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_, coef_ = perceptron(X,y)
```

```python
print(intercept_)
print(coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize = (10,6))
plt.plot(x_input,y_input,color = "red",linewidth = 3)
plt.scatter(X[:,0], X[:,1] ,c = y, cmap='winter', s = 100)
plt.ylim(-3,2)
```
